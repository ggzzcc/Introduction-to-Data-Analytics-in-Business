```python
import pandas as pd
```

```python
data = pd.read_csv('Multiple Cause of Death, 1999-2019 (2).txt',delimiter="\t")
data = data.iloc[:316,:]
data
```

```python
counties = {}
```

```python
for c in data['County']:
    if c not in counties:
        counties[c] = 1
    else:
        counties[c] += 1
```

```python
final_counties = [c for c in counties if counties[c] == 5]
```

```python
len(final_counties)
```

```python
county_filter = data['County'].apply(lambda x: x in final_counties)
```

```python
data[county_filter]
```

```python
data = data[county_filter]
```

```python
data = data.drop(['Notes','Year Code','County Code'], axis = 1)
```

```python
data
```

```python
data = data.reset_index(drop=True)
```

```python
data
```

```python
data.to_csv("data.csv")
```
