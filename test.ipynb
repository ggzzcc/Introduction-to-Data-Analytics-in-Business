```python
import pandas as pd
```

```python
edu_2015 = 'edu_2015.csv' # 88 county
data_file = 'data.xls'    # 50 county
```

```python
edu_2016 = 'edu_2016.csv' 
edu_2017 = 'edu_2017.csv' 
edu_2018 = 'edu_2018.csv' 
edu_2019 = 'edu_2019.csv' 
```

```python
df_2015 = pd.read_csv(edu_2015)
df_2016 = pd.read_csv(edu_2016)
df_2017 = pd.read_csv(edu_2017)
df_2018 = pd.read_csv(edu_2018)
df_2019 = pd.read_csv(edu_2019)
```

```python
df_2015
```

```python
def delete_ohio(x):  # for edu
    if x[-4:] == 'Ohio':
        return x[:-6]
    return x
```

```python
df_lists = [df_2015, df_2016, df_2017, df_2018, df_2019]
for df in df_lists:
    df['NAME'] = df['NAME'].apply(delete_ohio)
```

```python
def delete_OH(x):   # for data
    if x[-2:] == 'OH':
        return x[:-4]
    return x
```

```python
# read data
data = pd.read_excel(data_file)
data = data[data.County.notnull()]

# clean data
data['County'].apply(delete_OH)

data['County'] = data['County'].apply(delete_OH)
data = data.reset_index(drop=True)
```

```python
county = data.County.unique()
```

```python
cols = ['S1501_C01_002E',
        'S1501_C01_007E',
        'S1501_C01_008E']
```

```python
def my_fun(df):
    fileter = df.NAME.apply(lambda x: x in county)
    df = df[fileter]
    
    df = df[cols]
    df = df.reset_index(drop=True)
    df['S1501_C01_002E'] = df['S1501_C01_002E'].astype(int)
    df['S1501_C01_007E'] = df['S1501_C01_007E'].astype(int)
    df['S1501_C01_008E'] = df['S1501_C01_008E'].astype(int)
    df['sum'] = df.sum(axis=1) 
    return df
```

```python
new_list = []
for df in df_lists:
    new_list.append(my_fun(df))
```

```python
result = pd.concat(new_list)
result = result.reset_index(drop=True)
result
```

```python
data['education: less_than_high_school'] = result['sum']
```

```python
cols = ['Year', 'County', 'County Code', 'Deaths', 'Population', 'Crude Rate', '% of Total Deaths', 
        'Divorce Count', 'household: homeless', 'household: foster kids', 'household: people living alone',
       'education: less_than_high_school']
data = data[cols]
```

```python
data['Year'] = data['Year'].astype(int)
data['County Code'] = data['County Code'].astype(int)
data['Divorce Count'] = data['Divorce Count'].astype(int)
data['household: homeless'] = data['household: homeless'].astype(int)
data['household: foster kids'] = data['household: foster kids'].astype(int)
data['household: people living alone'] = data['household: people living alone'].astype(int)
```

```python
filename = 'combined_data.csv'
data.to_csv(filename, index=False)
```
