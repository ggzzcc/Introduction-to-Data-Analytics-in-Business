```python
def delete_ohio(x):
    if x[-4:] == 'Ohio':
        return x[:-6]
    return x
```

```python
def extract(file):
    
    def delete_ohio(x):
        if x[-4:] == 'Ohio':
            return x[:-6]
        return x
    
    df = pd.read_csv(file)
    df['NAME'] = df['NAME'].apply(delete_ohio)
    
    
    fileter = df.NAME.apply(lambda x: x in county) #county = df2.County.unique()
    final = df.copy()[fileter]
    final = final.iloc[:, :-1]
    
    cols = ['S1501_C01_002E','S1501_C01_007E','S1501_C01_008E']
    result = final[cols]
    result = result.reset_index(drop=True)
    
    result['S1501_C01_002E'] = result['S1501_C01_002E'].astype(int)
    result['S1501_C01_007E'] = result['S1501_C01_007E'].astype(int)
    result['S1501_C01_008E'] = result['S1501_C01_008E'].astype(int)
    result['sum'] = result.sum(axis=1) 
    
    return pd.DataFrame(result['sum'])
```

```python
extract('edu_2016.csv').head()
```

```python
extract('edu_2017.csv')
```

```python
def combine(files, answer=None):
    is_first = True
    for f in files:
        if is_first:
            is_first = False
            answer =extract(f)
        else:
            answer = pd.concat([answer, extract(f)])
    answer = answer.reset_index(drop=True)
    return answer
```
